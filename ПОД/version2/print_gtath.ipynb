```python
import matplotlib.pyplot as plt

#    
x_values = []
y_values = []
x_values_local = []
y_values_local = []

with open('/home/varvara//magistr_1sem//version2/data.txt', 'r') as file:
    for line in file:
        x, y = map(float, line.split())
        x_values.append(x)
        y_values.append(y)
        if (3 <= x <= 10):
            x_values_local.append(x)
            y_values_local.append(y)
        

#  
plt.figure(1)  #   
plt.plot(x_values, y_values, marker='o', linestyle='-', color='b')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('  Y  X ()')
plt.grid(True)
plt.show()

#  
plt.figure(2)  #   
plt.plot(x_values_local, y_values_local, marker='o', linestyle='-', color='r')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('  Y  X (  3<= x <= 10)')
plt.grid(True)
plt.show()
```
