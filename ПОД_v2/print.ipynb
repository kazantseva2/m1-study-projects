```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры
r0 = 2.94966 
A1, A0, p = 0.0611928, 0.0803404, 13.5777  # Параметры ERi
xi, q = 1.09285, 2.89005  # Параметры EBi

# A0 = 0.0992012 
# A1 = -0.0540063 
# p = 9.02197 
# q = 3.00369 
# xi = 1.1683 
# r0 = 2.84484

# Диапазон для расстояний r
r = np.linspace(2, 5, 100)  # Избегаем r = 0 для предотвращения деления на ноль

# Функции
def ERi(r):
    return (A1 * (r - r0) + A0) * np.exp(-p * ((r / r0) - 1))

def EBi(r):
    return (xi)* np.sqrt(np.exp(-2 * q * ((r / r0) - 1)))

# Общая энергия
E = ERi(r) - EBi(r)

# Построение графика
plt.figure(figsize=(8, 6))
plt.plot(r, E, label="E (Общая энергия)", linewidth=2)
plt.axhline(0, color='black', linewidth=0.8, linestyle=':')
plt.xlabel("Расстояние $r$")
plt.ylabel("Энергия")
plt.title("Энергия E как функция от $r$")
plt.legend()
plt.grid()
plt.show()
```
