```python
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('Task_10/2.csv', index_col=None)
columns = list(data.columns)
plt.figure(figsize=(5,5))
plt.scatter(data[columns[0]], data[columns[1]], c=data[columns[2]])
plt.show()

X = data[[columns[0], columns[1]]]
kmeans = KMeans(n_clusters=3, random_state=42).fit(X)

# Вывод результата
plt.figure(figsize=(5,5))
plt.scatter(data[columns[0]], data[columns[1]], c=kmeans.labels_)
plt.show()
```

```python
from sklearn.metrics import adjusted_rand_score

# Предположим, что у вас есть истинные метки 'true_labels' и предсказанные метки от KMeans 'predicted_labels'
true_labels = data[columns[2]]  # Истинные метки (нужны для оценки)
predicted_labels = kmeans.labels_  # Метки от KMeans

# Оценка качества кластеризации с помощью ARI
ari = adjusted_rand_score(true_labels, predicted_labels)
print(f"Adjusted Rand Index (ARI): {ari}")
```
